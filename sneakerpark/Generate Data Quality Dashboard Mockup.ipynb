```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({"userid": np.random.choice(["Good", "Bad"], 500),
             "email": np.random.choice(["Good", "Bad"], 500),
             "zipcode": np.random.choice(["Good", "Bad"], 500),
             "firstname": np.random.choice(["Good", "Bad"], 500)})
data
```

```python
data['firstname'] = 'Good'
fig, axes = plt.subplots(1, 4, sharey=True, figsize=(18, 6))

for i, col_name in enumerate(data.columns):
    counts = data[col_name].value_counts(normalize=True)
    try:
        counts['Bad']
    except KeyError as e:
        counts['Bad'] = 0.0
    counts.plot(ax=axes[i], kind='bar', color=['g', 'r'], rot=45)
    plt.yticks([])
    axes[i].text(-.075, counts['Good']-0.2, str(counts['Good']), color='w')
    axes[i].text(0.95, counts['Bad']-0.2, str(counts['Bad']), color='w')
    
axes[0].set_title("User ID in Customer Service\nRecords not in User System")
axes[3].set_title("First Names Present in Customer\nService Records not in User System")
axes[2].set_title("Invalid Email Address\nnot of Form __@__.com")
axes[1].set_title("Zip Codes with Length\nother Than 5")

plt.show()
```
